# **Introducción a Pandas**

Pandas es una librería de Python diseñada para **análisis y manipulación de datos**.  
Su nombre viene de *Panel Data* y se ha convertido en una de las herramientas más usadas en ciencia de datos.

## Objetos principales

- **Series**: estructura unidimensional (similar a un array con etiquetas).
- **DataFrame**: estructura bidimensional (tabla con filas y columnas).
- **Index**: etiquetas que identifican filas y columnas.

Pandas es como una **navaja suiza para datos**: permite leer, limpiar, transformar, analizar y visualizar información de manera sencilla y eficiente.


```python
import pandas as pd
```

### Diferencia entre `pd.Series` y `pd.DataFrame` en pandas

En **pandas**, una **Series** y un **DataFrame** son estructuras de datos distintas:

**Series**
- Es una estructura **unidimensional (1D)**.
- Representa una sola variable o columna.
- Tiene un **índice** y un conjunto de valores.
- Es equivalente a una columna de una tabla.

**DataFrame**
- Es una estructura **bidimensional (2D)**.
- Representa una **tabla de datos** con filas y columnas.
- Cada columna del DataFrame es una **Series**.
- Puede contener múltiples variables con distintos tipos de datos.

**Relación clave**
- `df["columna"]` devuelve una **Series**.
- `df[["columna"]]` devuelve un **DataFrame** con una sola columna.

**Resumen**
- Usa **Series** para trabajar con una sola variable.
- Usa **DataFrame** para analizar conjuntos de variables y sus relaciones.

## 5. Agrupación y agregación

```python
# Agrupación por Ciudad
print("\nAgrupación por Ciudad:\n", df.groupby("Ciudad")["Edad"].mean())

### Agrupación con `groupby`
"""
La función `groupby()` en Pandas permite agrupar filas de un DataFrame según los valores de una o más columnas, y luego aplicar operaciones de agregación sobre cada grupo.

#### Parámetros principales
- by (string o lista): columna(s) por las que se quiere agrupar.  
- aggfunc (función de agregación): operación que se aplica a cada grupo (ejemplo: `mean`, `sum`, `count`, etc.).

### Operaciones comunes con `groupby`

Después de agrupar con `df.groupby()`, se pueden aplicar distintas funciones de agregación:

- mean() → promedio de cada grupo  
- sum() → suma de valores  
- count() → número de elementos en cada grupo  
- min() / max() → mínimo y máximo de cada grupo  
- median() → mediana de cada grupo  
- std() / var() → desviación estándar y varianza  
- first() / last() → primer o último valor de cada grupo  
- agg() → aplicar múltiples funciones a la vez  
    Ejemplo: df.groupby("Ciudad")["Edad"].agg(["mean", "max", "min"])
"""
```

## 13. Funciones de Pivot y tablas

```python
# -------------------------------
# Ejemplo de tabla pivote en pandas
# -------------------------------

# Crear un DataFrame con información de ventas
df_pivot = pd.DataFrame({
    "Ciudad": ["La Paz", "La Paz", "Cochabamba", "Santa Cruz"],  # Categoría principal
    "Año": [2023, 2024, 2023, 2024],                            # Variable que se convertirá en columnas
    "Ventas": [100, 150, 200, 250]                              # Valores a agregar
})
print("\nDataFrame original:\n", df_pivot)

# Crear la tabla pivote
# values  : columna numérica que se va a resumir
# index   : filas de la tabla pivote
# columns : columnas de la tabla pivote
# aggfunc : función de agregación (por defecto es la media)
tabla_pivote = pd.pivot_table(
    df_pivot,
    values="Ventas",
    index="Ciudad",
    columns="Año"
)

# Mostrar el resultado
print("\nTabla pivote:\n", tabla_pivote)
```

## 14. Funciones de ventana

```python
# Media móvil (rolling mean)

# Crear un DataFrame numérico de ejemplo
df_num = pd.DataFrame({
    "Valores": [9, 22, 31, 47, 59, 60]
})

# Calcular la media móvil
# window=3 indica que el promedio se calcula usando
# los últimos 3 valores consecutivos
media_movil = df_num["Valores"].rolling(window=3).mean()

# Mostrar el resultado
print("\nMedia móvil:\n", media_movil)
```

```python
# Suma acumulada (cumulative sum)
print("\nSuma acumulada:\n", df_num["Valores"].cumsum())
```
