# **Introducción a Numpy**

**NumPy** (Numerical Python) es una librería fundamental para el cálculo científico en Python.  
Proporciona soporte para **arrays multidimensionales** y una amplia colección de funciones matemáticas de alto rendimiento.

## Características principales
- **Arrays (ndarray):** estructuras eficientes para almacenar y manipular datos numéricos.
- **Operaciones vectorizadas:** permiten cálculos rápidos sin necesidad de bucles explícitos.
- **Funciones matemáticas avanzadas:** trigonometría, álgebra lineal, estadísticas, transformadas de Fourier.
- **Manejo de datos multidimensionales:** matrices, tensores y operaciones sobre ellos.
- **Interoperabilidad:** se integra fácilmente con otras librerías como Pandas, Matplotlib y Scikit-learn.

## Ventajas
- Mucho más rápido y eficiente que las listas nativas de Python para cálculos numéricos.
- Sintaxis clara y concisa para operaciones matemáticas.
- Base sobre la que se construyen muchas librerías de ciencia de datos y aprendizaje automático.

NumPy es el **pilar del ecosistema científico en Python**, ofreciendo herramientas poderosas para trabajar con datos numéricos y realizar cálculos de manera eficiente.

# Distribuciones en `np.random` (NumPy)

El módulo `numpy.random` permite generar números aleatorios siguiendo distintas **distribuciones de probabilidad**.  
Estas funciones son muy útiles en simulaciones, estadística y machine learning.

## Distribuciones principales

### Uniforme
- **`np.random.uniform(low, high, size)`**
- Genera valores dentro de un rango `[low, high)` con igual probabilidad.

### Normal (Gaussiana)
- **`np.random.normal(loc, scale, size)`**
- Distribución de campana centrada en `loc` (media) con desviación estándar `scale`.

### Binomial
- **`np.random.binomial(n, p, size)`**
- Número de éxitos en `n` ensayos con probabilidad `p` de éxito.

### Poisson
- **`np.random.poisson(lam, size)`**
- Modela el número de eventos que ocurren en un intervalo dado con tasa `lam`.

### Exponencial
- **`np.random.exponential(scale, size)`**
- Tiempo entre eventos en un proceso de Poisson, con parámetro `scale`.

### Beta
- **`np.random.beta(a, b, size)`**
- Distribución continua en el rango `[0,1]`, usada en estadística bayesiana.

### Gamma
- **`np.random.gamma(shape, scale, size)`**
- Generalización de la exponencial, útil en tiempos de espera y colas.

### Chi-cuadrado
- **`np.random.chisquare(df, size)`**
- Distribución usada en pruebas de hipótesis, con `df` grados de libertad.

### Lognormal
- **`np.random.lognormal(mean, sigma, size)`**
- Valores positivos cuya log-transformación sigue una distribución normal.

### Geométrica
- **`np.random.geometric(p, size)`**
- Número de ensayos hasta el primer éxito con probabilidad `p`.

### Multinomial
- **`np.random.multinomial(n, pvals, size)`**
- Generalización de la binomial para múltiples categorías.

### F-distribution
- **`np.random.f(dfnum, dfden, size)`**
- Usada en pruebas de varianza y ANOVA.

Estas funciones son esenciales para **simulación estadística y modelado probabilístico** en Python.

```python
# Fijar semilla para reproducibilidad
np.random.seed(42)

# Número de registros simulados
n = 100

# Crear DataFrame simulado
df = pd.DataFrame({
    "Precio_Accion": np.random.uniform(50, 500, n),       # Precios de acciones entre 50 y 500
    "Volumen": np.random.randint(1000, 100000, n),        # Volumen de transacciones
    "Retorno": np.random.normal(0, 0.05, n),             # Retornos diarios, distribución normal
    "Riesgo": np.random.uniform(0.01, 0.3, n),           # Riesgo asociado a la acción
    "Dividendos": np.random.choice([0, 0.01, 0.02, 0.03], n),  # Dividendos posibles
    "Beta": np.random.uniform(0.5, 1.5, n),              # Beta del activo
    "Ratio_P_E": np.random.uniform(5, 40, n),            # Ratio Precio/Utilidad
    "Inflacion": np.random.normal(0.02, 0.01, n),        # Inflación diaria simulada
})

# Mostrar las primeras filas
print(df.head())
```

```python
# -------------------------------
# Creación de arrays NumPy desde columnas del DataFrame
# -------------------------------

# Convertir columnas de DataFrame en arrays NumPy
# Esto permite aplicar operaciones vectorizadas, álgebra lineal y funciones matemáticas de manera eficiente
arr_precio = df["Precio_Accion"].to_numpy()  # Alternativa: np.array(df["Precio_Accion"])
arr_volumen = df["Volumen"].to_numpy()       # Mantener consistencia y eficiencia

# Verificar los primeros elementos para confirmar la conversión
print("Primeros 5 valores de Precio_Accion:", arr_precio[:5])
print("Primeros 5 valores de Volumen:", arr_volumen[:5])
```

```python
# -------------------------------
# Arrays especiales: ceros, unos y vacíos
# -------------------------------

# Array de ceros: útil para inicializar vectores o acumuladores
zeros_array = np.zeros(5)  # Crea un array de 5 elementos todos con valor 0
print("Array de ceros:", zeros_array)

# Array de unos: útil para inicializar multiplicadores o máscaras
ones_array = np.ones(5)    # Crea un array de 5 elementos todos con valor 1
print("Array de unos:", ones_array)

# Array vacío: crea un array sin inicializar (valores aleatorios de memoria)
# Útil cuando se va a rellenar inmediatamente con datos específicos
empty_array = np.empty(5)
print("Array vacío (valores no inicializados):", empty_array)

# Array con un valor específico
full = np.full((2,3), 7)      # [[7 7 7] [7 7 7]]
print("Array lleno con un valor específico:\n", full)
```

```python
# -------------------------------
# Arrays con rangos
# -------------------------------

# np.arange(start, stop, step): crea un array con valores desde start hasta stop (excluido) con paso step
arange_array = np.arange(0, 50, 5)  # 0, 5, 10, ..., 45
print("Array con arange:", arange_array)

# np.linspace(start, stop, num): crea un array de 'num' valores igualmente espaciados entre start y stop (incluido)
linspace_array = np.linspace(0, 1, 5)  # 0.0, 0.25, 0.5, 0.75, 1.0
print("Array con linspace:", linspace_array)

# np.logspace(start, stop, num): crea un array de 'num' valores igualmente espaciados en escala logarítmica
# start y stop son exponentes de base 10 por defecto
logspace_array = np.logspace(0, 2, 5)  # 10^0, 10^0.5, 10^1, 10^1.5, 10^2
print("Array con logspace:", logspace_array)
```

```python
# -----------------------------------------
# Funciones matemáticas y operaciones con NumPy
# -----------------------------------------

# Convertimos la columna Precio_Accion a un arreglo NumPy
x = df["Precio_Accion"].values  

# Convertimos la columna Volumen a un arreglo NumPy
y = df["Volumen"].values       

# 1. Suma elemento a elemento (Precio_Accion + Volumen)
suma = np.add(x, y)           
df["Suma"] = suma              # Guardamos la suma en una nueva columna

# 2. Resta elemento a elemento (Precio_Accion - Volumen)
resta = np.subtract(x, y)     
df["Resta"] = resta            # Guardamos la resta en una nueva columna

# 3. Multiplicación elemento a elemento (Precio_Accion * Volumen)
producto = np.multiply(x, y)  
df["Producto"] = producto      # Guardamos el producto en una nueva columna

# 4. División elemento a elemento (Precio_Accion / Volumen)
division = np.divide(x, y)     
df["Division"] = division      # Guardamos la división en una nueva columna

# 5. Potencia (Precio_Accion elevado al cuadrado)
potencia = np.power(x, 2)      
df["Precio_Potencia2"] = potencia  # Guardamos la potencia en una nueva columna

# 6. Módulo (Volumen mod Precio_Accion)
# Calcula el resto de la división de cada elemento de a entre el correspondiente elemento de b.
mod = np.mod(y, x)             
df["Modulo"] = mod             # Guardamos el resultado del módulo en una nueva columna

# 7. Resto (remainder, similar a módulo pero con manejo distinto de negativos)
resto = np.remainder(y, x)    
df["Resto"] = resto            # Guardamos el resto en una nueva columna

# 8. División entera (floor divide, división redondeada hacia abajo)
div_entera = np.floor_divide(y, x)  
df["Div_Entera"] = div_entera  # Guardamos la división entera en una nueva columna

# Mostrar las primeras filas del DataFrame con todas las operaciones
print("\nDataFrame con operaciones NumPy:\n", df.head())
```

```python
# -----------------------------------------
# Funciones para encontrar índices de valores mínimos y máximos
# -----------------------------------------

# Convertir columna a array NumPy
precios = df["Precio_Accion"].values

# Índice del precio mínimo
idx_min = np.argmin(precios)
print("\nÍndice del precio mínimo:", idx_min)
print("Precio mínimo:", precios[idx_min])

# Índice del precio máximo
idx_max = np.argmax(precios)
print("\nÍndice del precio máximo:", idx_max)
print("Precio máximo:", precios[idx_max])
```

```python
# -------------------------------
# Matrices identidad y diagonal
# -------------------------------

# np.identity(n): crea una matriz identidad cuadrada de tamaño n x n
# Matriz identidad: 1 en la diagonal principal y 0 en el resto
identity_matrix = np.identity(3)
print("Matriz identidad 3x3:\n", identity_matrix)
```

```python
# np.eye(N, M=None, k=0): crea una matriz con 1 en la diagonal k y 0 en el resto
# N: número de filas, M: número de columnas (opcional), k: desplazamiento de la diagonal
eye_matrix = np.eye(3, 4, k=1)  # 3 filas, 4 columnas, diagonal desplazada una posición hacia arriba
print("\nMatriz con diagonal desplazada k=1 (3x4):\n", eye_matrix)
```

```python
# -------------------------------
# Comparaciones
# -------------------------------
print("Comparaciones element-wise:")
print("Precio > 300:\n", np.greater(df["Precio_Accion"], 300).value_counts(), "\n")
print("Riesgo < 0.1:\n", np.less(df["Riesgo"], 0.1).value_counts(), "\n")
print("Dividendos == 0.02:\n", np.equal(df["Dividendos"], 0.02).value_counts(), "\n")
print("Retorno != 0:\n", np.not_equal(df["Retorno"], 0).value_counts(), "\n")
```

```python
# -------------------------------
# Operaciones lógicas con NumPy
# -------------------------------

print("Operaciones lógicas:")

# Crear máscaras booleanas según condiciones financieras
# np.greater: devuelve True si el valor es mayor que el límite
alta_rentabilidad = np.greater(df["Retorno"], 0.02)  # Retorno > 2%

# np.less: devuelve True si el valor es menor que el límite
bajo_riesgo = np.less(df["Riesgo"], 0.1)  # Riesgo < 0.1

# np.logical_and: True si ambas condiciones se cumplen
print("Alta rentabilidad AND bajo riesgo:\n", np.logical_and(alta_rentabilidad, bajo_riesgo).value_counts(), "\n")

# np.logical_or: True si al menos una de las condiciones se cumple
print("Alta rentabilidad OR bajo riesgo:\n", np.logical_or(alta_rentabilidad, bajo_riesgo).value_counts(), "\n")

# np.logical_not: invierte los valores booleanos (True -> False, False -> True)
print("NOT bajo riesgo:\n", np.logical_not(bajo_riesgo).value_counts(), "\n")
```

```python
# -------------------------------
# Condicional con np.where
# -------------------------------
df["Clasificacion"] = np.where(
    np.logical_and(alta_rentabilidad, bajo_riesgo),
    "Premium",
    np.where(alta_rentabilidad, "Rentable", "Normal")
)
print("Clasificación de activos:\n", df[["Retorno", "Riesgo", "Clasificacion"]].head(), "\n")

print(df['Clasificacion'].value_counts())
```

```python
# -------------------------------
# any y all
# -------------------------------
print("any y all:")
print("¿Algún activo con retorno > 0.04?", np.any(df["Retorno"] > 0.04))
print("¿Todos los activos tienen riesgo < 0.3?", np.all(df["Riesgo"] < 0.3))
```

```python
# -------------------------------
# Transformaciones de arrays: reshape, ravel, flatten
# -------------------------------

# - `reshape(filas, columnas)`: cambia la **forma** del array sin aplanarlo.  
# - `ravel()`: aplana el array en 1D y devuelve una **vista** (modificarlo puede afectar el original).  
# - `flatten()`: aplana el array en 1D y devuelve una **copia** (no afecta el original).

# Queremos reorganizar arr_precio en 10 filas
filas = 10
columnas = 10  # Para arr_precio de 100 elementos, columnas = 100/10 = 10

# -------------------------------
# Reshape: convierte un array 1D en una matriz 2D
# La nueva matriz tendrá 'filas' filas y 'columnas' columnas
# Debe coincidir el número total de elementos
reshaped = arr_precio.reshape(filas, columnas)
print("Array reshaped (10 filas x 10 columnas):\n", reshaped)

# -------------------------------
# Ravel: aplana la matriz en un array 1D
# Devuelve una **vista**, por lo que modificar raveled puede afectar arr_precio original
raveled = reshaped.ravel()
print("\nArray raveled (aplanado, vista 1D):\n", raveled)

# -------------------------------
# Flatten: aplana la matriz en un array 1D
# Devuelve una **copia**, por lo que modificar flattened no afecta arr_precio original
flattened = reshaped.flatten()
print("\nArray flattened (aplanado, copia):\n", flattened)
```

```python
# -------------------------------
# Transpose, swapaxes, moveaxis (mejorado)
# -------------------------------

# Crear matriz 2D con precios y volúmenes
matrix_2d = np.array([arr_precio, arr_volumen])
print("Matriz 2D original (2 filas x 10 columnas):\n", matrix_2d, "\n")

# Transpuesta: filas ↔ columnas
transposed = matrix_2d.T
print("Transpuesta (10 filas x 2 columnas):\n", transposed, "\n")

# Swapaxes: intercambiar ejes 0 y 1 (equivalente a transpose en 2D)
swapped = np.swapaxes(matrix_2d, 0, 1)
print("Swapaxes (ejes 0 y 1 intercambiados):\n", swapped, "\n")
```

```python
# -------------------------------
# Ejemplo 3D para moveaxis
# -------------------------------

# Crear un array 3D a partir de una matriz 2D
# matrix_2d tiene shape (2,10)
# Creamos 2 "capas": la original y la original + 1000
matrix_3d = np.array([matrix_2d, matrix_2d + 1000])  # shape (2, 2, 10)
print("Matriz 3D original (2 capas x 2 filas x 10 columnas):\n", matrix_3d, "\n")

# np.moveaxis: mueve el eje especificado a otra posición
# Aquí movemos el eje 0 (capas) al eje 2 (columnas)
# Antes: (capas=2, filas=2, columnas=10)
# Después: (filas=2, columnas=10, capas=2)
moved = np.moveaxis(matrix_3d, 0, 2)
print("Moveaxis (eje 0 movido a eje 2) - shape:", moved.shape, "\n")
print(moved, "\n")
```

```python
# -------------------------------
# Manipulación avanzada de arrays: concatenate, stack, split, tile, repeat
# -------------------------------

# Concatenar arrays: une arr_precio y arr_volumen en un solo array 1D
concatenated = np.concatenate([arr_precio, arr_volumen])
print("Array concatenado (1D):\n", concatenated)
```

```python
# Stack vertical: apila arrays como filas de una nueva matriz 2D
stacked_v = np.vstack([arr_precio, arr_volumen])
print("\nStack vertical (vstack, filas):\n", stacked_v)
```

```python
# Stack horizontal: apila arrays como columnas de una nueva matriz 2D
stacked_h = np.hstack([arr_precio, arr_volumen])
print("\nStack horizontal (hstack, columnas):\n", stacked_h)
```

```python
# Split: divide arr_precio en 2 partes iguales (lista de arrays)
split_array = np.split(arr_precio, 2)
print("\nArray dividido en 2 partes:")
for i, part in enumerate(split_array, start=1):
    print(f"Parte {i}:", part)
```

```python
# Tile: repite la secuencia completa de arr_precio 2 veces
tiled_array = np.tile(arr_precio, 2)
print("\nArray con secuencia repetida (tile 2 veces):\n", tiled_array)
```

```python
# Repeat: repite **cada elemento** de arr_precio 2 veces
repeated_array = np.repeat(arr_precio, 2)
print("\nArray con cada elemento repetido 2 veces (repeat):\n", repeated_array)
```

```python
# -------------------------------
# Operaciones sobre arrays aplicadas al DataFrame
# -------------------------------

# Suma acumulativa de Dividendos
df["Cumsum_Dividendos"] = np.cumsum(df["Dividendos"])

# Diferencias entre precios consecutivos de la acción
df["Diff_Precio"] = np.diff(df["Precio_Accion"], prepend=df["Precio_Accion"].iloc[0])

# Redondeo del Ratio Precio/Utilidad a 2 decimales
df["Ratio_P_E_Round"] = np.round(df["Ratio_P_E"], 2)

# Mostrar las primeras filas para verificar
print(df[["Dividendos", "Cumsum_Dividendos", "Precio_Accion", "Diff_Precio", "Ratio_P_E", "Ratio_P_E_Round"]].head())
```

```python
# -------------------------------
# Operaciones lógicas y condicionales
# -------------------------------

# Crear columna booleana indicando si el Volumen es alto (>50,000)
df["Alta_Volumen"] = df["Volumen"] > 50000

# Crear columna categórica según el Retorno
df["Estado_Retorno"] = np.where(df["Retorno"] > 0.02, "Ganancia", "Perdida")

# Mostrar las primeras filas para verificar
print(df[["Volumen", "Alta_Volumen", "Retorno", "Estado_Retorno"]].head())
```

```python
# -------------------------------
# Transformaciones matemáticas avanzadas
# -------------------------------

# Redondear Riesgo a 2 decimales
df["Riesgo_Round"] = np.round(df["Riesgo"], 2)

# Piso y techo de Riesgo
df["Riesgo_Floor"] = np.floor(df["Riesgo"])
df["Riesgo_Ceil"] = np.ceil(df["Riesgo"])

# Valor absoluto del Retorno
df["Retorno_Abs"] = np.abs(df["Retorno"])

# Raíz cuadrada del Volumen (asegurando valores positivos)
df["Sqrt_Volumen"] = np.sqrt(df["Volumen"].clip(lower=0))

# Mostrar las primeras filas para verificar
print(df[["Riesgo", "Riesgo_Round", "Riesgo_Floor", "Riesgo_Ceil", "Retorno", "Retorno_Abs", "Volumen", "Sqrt_Volumen"]].head())
```

```python
# -------------------------------
# Operaciones matriciales y álgebra lineal
# -------------------------------

# Convertir columnas seleccionadas en matriz NumPy
matriz_precios = df[["Precio_Accion", "Beta", "Ratio_P_E"]].values
```

```python
# Producto matricial (X^T * X)
producto = np.dot(matriz_precios.T, matriz_precios)
print("Producto matricial (X^T * X):\n", producto)
```

```python
# Determinante con pequeña regularización para evitar singularidades
determinante = np.linalg.det(producto + np.eye(producto.shape[0]) * 1e-6)
print("\nDeterminante con regularización:", determinante)
```

```python
# -------------------------------
# Eigenvalores y eigenvectores con explicación
# -------------------------------

# Calcular eigenvalores y eigenvectores de la matriz 'producto' (X^T * X)
# Eigenvalores (λ): indican la cantidad de varianza explicada por cada componente principal.
#                  Valores más grandes = más importancia de esa dirección en los datos.
# Eigenvectores (v): vectores que indican la dirección de cada componente principal en el espacio de variables.
eigenvalores, eigenvectores = np.linalg.eig(producto)

# Ordenar los eigenvalores de mayor a menor para identificar las componentes más importantes
orden = np.argsort(eigenvalores)[::-1]  # argsort devuelve los índices que ordenan el array
eigenvalores_ordenados = eigenvalores[orden]  # Eigenvalores ordenados
eigenvectores_ordenados = eigenvectores[:, orden]  # Reorganizar columnas de eigenvectores según el orden

# Mostrar eigenvalores de manera legible con 4 decimales
print("\nEigenvalores ordenados de mayor a menor:")
for i, val in enumerate(eigenvalores_ordenados, start=1):
    print(f"λ{i}: {val:.4f}  # Varianza explicada por esta componente")

# Mostrar eigenvectores correspondientes a cada eigenvalor
print("\nEigenvectores correspondientes (columnas representan vectores):")
for i in range(eigenvectores_ordenados.shape[1]):
    # Cada columna del eigenvector corresponde a la dirección principal de la varianza de los datos
    print(f"v{i+1}: {eigenvectores_ordenados[:, i]}")
```
